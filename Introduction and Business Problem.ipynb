# The Battle of Neighbourhoods (Week 1)

My client is looking to open a Japanese restaurant in North York, Toronto. He would like the restaurant to be located in a neighbourhood of North York (e.g. Parkwoods, Victoria Village, etc.) that has the fewest number of restaurants as a common venue.

The reasoning for his decision is that, a neighbourhood with fewer restaurants will attract more customers to his restaurants, therefore bringing in more businesses.

## Data Requirements

### 1. I will need a list of neighbourhoods in North York.
* I will obtain a list of neighbourhoods in North York through this URL: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M 
* I will create a dataframe containing a list of neighbourhoods in North York, as well as their postal codes, latitudes, and longitudes.

### 2. I will need to explore and analyze neighbourhoods in North York.
* I will use Foursquare data to segment and cluster neighbourhoods in North York.
* In particular, I will find out the most common venues of each neighbourhood in North York.
* For example, I want to find out if Parkwoods or Victoria Village would have "cafes", "deli", "Italian restaurants", or "doughnut shops" listed as their top 10 most common venues.

### 3. Gathering data and Determining Business Action
* Through the above steps, I should be able to obtain a list of top 10 most common venues for all neighbourhoods in North York.
* I can then count how many times an eatery (e.g. "cafes", "deli", "Italian restaurants", or "doughnut shops") appears as a most common venue for each neighbourhood.
* The best neighbourhood for my client to open his Japanese restaurant would be the neighbourhood with the least count of an eatery as a most common venue.
